### Fazendo o tratamento das variáveis

```python
# Fazendo uma cópia dos dados em outra variável
# Por conter muitos dados missing a feature Cabin não foi incluída
df1 = df[['Pclass', 'Sex', 'SibSp', 'Parch', 'Age', 'Name', 'Fare', 'Embarked', 'Survived']].copy()
```

```python
# Criando a coluna Saudacao e IdadeSaudacao para preencher os dados faltantes na coluna Age
df1['Saudacao'] = df1.Name.apply(lambda palavra: palavra.split(',')[1].split(' ')[1])

saudacoes_possiveis = df1.Saudacao.value_counts()[:6]
saudacoes_possiveis = saudacoes_possiveis.index

df1.Saudacao = df1.Saudacao.apply(lambda x: x if x in saudacoes_possiveis else 'Outros')

idade_saudacao = df1.groupby('Saudacao').agg({'Age': 'mean'}).reset_index()
idade_saudacao = idade_saudacao.rename({'Age': 'IdadeSaudacao'}, axis = 1)

df1 = df1.merge(idade_saudacao)

df1.Age = df1.Age.combine_first(df1.IdadeSaudacao)

df1.Sex = df1.Sex.map({'male': 1, 'female': 0})
```

```python
# Podemos observar que os dados na coluna Age já foram preenchidos de acordo com a idadesaudacao
msno.matrix(df1, color=(.07,.03,.5));
```

```python
# Retirando as colunas Name e IdadeSaudacao
df1 = df1.drop(['Name', 'IdadeSaudacao'], axis = 1)
```

```python
df1
```

```python
df1.Saudacao.value_counts()
```

```python
# Categorizando a variável Saudacao
def trata_saudaçao(valor):
    if valor in ['Mr.']:
        return 1
    elif valor in ['Miss.']:
        return 2
    elif valor in ['Mrs.']:
        return 3
    elif valor in ['Master.']:
        return 4
    elif valor in ['Outros']:
        return 5
    elif valor in ['Dr.']:
        return 6    
    else:
        return 7
    
df1['Saudacao'] = df1['Saudacao'].apply(lambda x: trata_saudaçao(x))
```

```python
# Categorizando a variável Embarked
def trata_embarked(valor):
    if valor in ['S']:
        return 1
    elif valor in ['C']:
        return 2
    else:
        return 3
    
df1['Embarked'] = df1['Embarked'].apply(lambda x: trata_embarked(x))
```

### Agora iremos tentar balancear as classes

```python
# A classe 0 esta bem superior a classe 1
df1.Survived.value_counts(1)
```

```python
df1
```

```python
# Undersampling – Consiste em reduzir de forma aleatória os exemplos da classe majoritária.
def UnderSampling(df1, under = 1):
    classe0 = df1[df1['Survived'] == 0]
    classe1 = df1[df1['Survived'] == 1]
    
    amostra0 = classe0.sample(under * classe1.shape[0])
    under = pd.concat([amostra0, classe1])
    
    return under
```

```python
under = UnderSampling(df1, 1)
print(under.shape)
```

```python
# Conseguimos igualar as classes
under.Survived.value_counts(1)
```

```python
under
```

```python
# Rodaremos os mesmos modelos agora para a classe balanceada
modelagem = setup(data = under, 
                 target = 'Survived', 
                 test_data = df_teste,
                 normalize = True)

modelo = compare_models(include = modelos, sort = 'F1')
```

```python
pred1 = predict_model(modelo)
```

```python
# Tentaremos fazer o Under e Oversapling juntos agora para aumentar os dados da classe 1
def UnderOverSampling(df, under = 1, over = 1):
    classe0 = df[df['Survived'] == 0]
    classe1 = df[df['Survived'] == 1]
    
    amostra0 = classe0.sample(under * classe1.shape[0])
    amostra1 = pd.concat([classe1] * over, ignore_index=True)
    tudo = pd.concat([amostra0, amostra1])
    
    return tudo
```

```python
underover1 = UnderOverSampling(df1, under = 1, over = 2)
print(underover1.shape)
```

```python
# Classe 1 aumentada
underover1.Survived.value_counts(1)
```

```python
underover1
```

```python
modelagem = setup(data = underover1, 
                 target = 'Survived', 
                 test_data = df_teste,
                 normalize = True)

modelo = compare_models(include = modelos, sort = 'F1')
```

```python
pred = predict_model(modelo)
```
